```python
import sys
import string
import matplotlib.pyplot as plt
```

<h2>Mehak Sharma | BioE 131 | Fall 2019 </h2>

```python
text = open("romeoandjuliet.txt")

counts = dict()
for line in text:
    line = line.lower()
    words = [word.strip(string.punctuation) for word in line.split()]
    for word in words:
        if word not in counts:
            counts[word] = 1
        else:
            counts[word] += 1

print(counts)
```

```python
words = list(counts.keys())
values = list(counts.values())

plt.bar(range(len(counts)),values,tick_label=words)
plt.xlabel("Words")
plt.ylabel("Counts")
plt.title("Word Count values in Shakespeare's Romeo and Juliet")
plt.show()
```

<h4>Ten Most Frequent Words </h4>

```python
import operator
dict(sorted(counts.items(), key=operator.itemgetter(1), reverse=True)[:10])
```

The ten most frequent words are articles (such as  "the" and "a"), conjunctions ("and"), prepositions ("in", "to", "of"), pronouns ("I", "my"). This makes sense because words are connector phrases that are used a lot in everyday speech.

<h4>Ten Least Frequent Words </h4>

```python
dict(sorted(counts.items(), key=operator.itemgetter(1), reverse=False)[:10])
```

These words were used least, and most of them are unsurprising except for "tragedy" and "family".
